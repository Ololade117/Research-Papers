R1.3 Write a short Python function, minmax(data), that takes a sequence of
one or more numbers, and returns the smallest and largest numbers, in the
form of a tuple of length two. Do not use the built-in functions min or
max in implementing your solution.

```python
def minmax(data:list[int]) -> int:
  a = data[0]
  b = data[0]
  for n in data:
    if n < a:
      a = n
    elif n > b:
      b = n
  minimax = (a, b)
  return minimax

#test
minmax([1, 0, 1, 2, 10])
```

R 1.4 Write a short Python function that takes a positive integer n and returns
the sum of the squares of all the positive integers smaller than n.

```python
def sum_of_squares(n:int) -> int:
  sums = 0
  for i in range(n):
    sums += i**2
  return sums

sum_of_squares(5)
```

R 1.6 Write a short Python function that takes a positive integer n and returns

```python
def odd_sum(n:int) -> int:
  sums = 0
  for i in range(1, n, 2):
    sums += i**2

  return sums

odd_sum(5)
```

R 1.2 Python’s random module includes a function choice(data) that returns a
random element from a non-empty sequence. The random module in
cludes a more basic function randrange, with parameterization similar to
the built-in range function, that return a random choice from the given
range. Using only the randrange function, implement your own version
of the choice function.

```python
import random
def choice(data:list[int]) -> int:
  idx = random.randrange(len(data))
  return data[idx]

choice([1, 0, -2, 3, 6, 81, 9])
```

C 1.14 Write a short Python function that takes a sequence of integer values and
determines if there is a distinct pair of numbers in the sequence whose
product is odd.

```python
def is_pair_odd(data : list) -> bool:
  count = 0
  for i in data:
    if i%2 ==1:
      count += 1
  return bool((count-1))

print(is_pair_odd([1,2,4,7,5]))
```

C 1.15 Write a Python function that takes a sequence of numbers and determines
if all the numbers are different from each other (that is, they are distinct).

```python
def distinct(data:list) -> bool:
  return len(data) == len(set(data))

distinct([1,2,3,4,5])
```

C 1.21 Write a Python program that repeatedly reads lines from standard input
until an EOFError is raised, and then outputs those lines in reverse order
(a user can indicate end of input by typing ctrl-D).

```python
user_inputs = []
while True:
  try:
    user_input = input("Enter Something:")
    user_inputs.append(user_input)
  except EOFError:
    break


user_inputs.reverse()
print("".join(user_inputs))
```
