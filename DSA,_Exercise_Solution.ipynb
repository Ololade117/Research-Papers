R1.3 Write a short Python function, minmax(data), that takes a sequence of
one or more numbers, and returns the smallest and largest numbers, in the
form of a tuple of length two. Do not use the built-in functions min or
max in implementing your solution.

```python
def minmax(data:list[int]) -> int:
  a = data[0]
  b = data[0]
  for n in data:
    if n < a:
      a = n
    elif n > b:
      b = n
  minimax = (a, b)
  return minimax

#test
minmax([1, 0, 1, 2, 10])
```

R 1.4 Write a short Python function that takes a positive integer n and returns
the sum of the squares of all the positive integers smaller than n.

```python
def sum_of_squares(n:int) -> int:
  sums = 0
  for i in range(n):
    sums += i**2
  return sums

sum_of_squares(5)
```

C 1.22 Write a short Python program that takes two arrays a and b of length n
storing int values, and returns the dot product of a and b. That is, it returns
an array c of length n such that c[i]= a[i] · b[i], for i = 0,...,n−1.

```python
def dot_product(a: list, b:list) -> float|str:
  if len(a) != len(b):
    raise ValueError("Impossible to compute dot product")

  total = 0

  for i in range(len(a)):
    total += a[i]*b[i]

  return total

dot_product([1,2,3], [1, 2, 3])
```

C 1.24 Write a short Python function that counts the number of vowels in a given
character string.

```python
def vowel_counter(a:str) -> int:
  vowels = ["a", "e", "i", "o", "u"]
  counts = 0
  for n in a:
    if n in vowels:
      counts += 1
  return counts

vowel_counter("abcdef")
```

C 1.27 In Section 1.8, we provided three different implementations of a generator
that computes factors of a given integer. The third of those implementations, from page 41, was the most efficient, but we noted that it did not
yield the factors in increasing order. Modify the generator so that it reports
factors in increasing order, while maintaining its general performance advantages.


```python
def factors(n): # generator that computes factors
  k=1
  while k *k < n: # while k < sqrt(n)
    if n % k == 0:
      yield k
      yield n // k
  k += 1
  if k* k == n: # special case if n is perfect square
    yield k

factors(5)
```
